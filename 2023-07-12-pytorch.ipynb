```python
from torch import sqrt, pow, tensor, zeros
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# Create a 600x600 tensor filled with zeros
image = zeros((600, 600))

center_x = image.size(1) // 2
center_y = image.size(0) // 2
radius = 200
for i in range(image.size(0)):
    for j in range(image.size(1)):
        distance = sqrt(pow(tensor(i) - center_y, 2) + pow(tensor(j) - center_x, 2))
        if distance <= radius:
            image[i, j] = 1

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
image = zeros((600, 600))
center_x = image.size(1) // 2
center_y = image.size(0) // 2
radius = 200
radius2 = 100
for i in range(image.size(0)):
    for j in range(image.size(1)):
        distance = sqrt(pow(tensor(i) - center_y, 2) + pow(tensor(j) - center_x, 2))
        if distance <= radius2:
            image[i, j] = 1
        elif distance <= radius:
            image[i, j] = 0.5

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
image = zeros((600, 600))
center_x = image.size(1) // 2
center_y = image.size(0) // 2
radii = [360, 200, 100, 50]
for i in range(image.size(0)):
    for j in range(image.size(1)):
        distance = sqrt(pow(tensor(i) - center_y, 2) + pow(tensor(j) - center_x, 2))
        for rindex, radius in enumerate(radii):
            if distance <= radius:
                image[i, j] = (rindex + 1) / len(radii)

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
image = zeros((600, 600))

# Draw a diagonal line across the image
for i in range(image.size(0)):
    for j in range(image.size(1)):
        if i == j:
            image[i, j] = 1

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def draw_fractal(image, top_left, bottom_right):
    # Base case: if the width or height of the region is less than 3 pixels, return
    if bottom_right[0] - top_left[0] <= 2 or bottom_right[1] - top_left[1] <= 2:
        return

    # Draw a diagonal line in the center of the region
    mid_x = (top_left[0] + bottom_right[0]) // 2
    mid_y = (top_left[1] + bottom_right[1]) // 2
    for i in range(top_left[0], bottom_right[0]):
        for j in range(top_left[1], bottom_right[1]):
            if i == j or (i == mid_x and j >= mid_y) or (j == mid_y and i >= mid_x):
                image[i, j] = 1

    # Recursively divide the region into four quadrants
    draw_fractal(image, top_left, (mid_x, mid_y))
    draw_fractal(image, (top_left[0], mid_y + 1), (mid_x, bottom_right[1]))
    draw_fractal(image, (mid_x + 1, top_left[1]), (bottom_right[0], mid_y))
    draw_fractal(image, (mid_x + 1, mid_y + 1), bottom_right)

image = zeros((600, 600))

draw_fractal(image, (0, 0), (599, 599))

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
BASE_CASE_SIZE = 3

def draw_fractal(image, top_left, bottom_right):
    # Base case: if the width or height of the region is less than the base case size, return
    if bottom_right[0] - top_left[0] + 1 < BASE_CASE_SIZE or bottom_right[1] - top_left[1] + 1 < BASE_CASE_SIZE:
        return

    # Draw a diagonal line in the center of the region
    mid_x = (top_left[0] + bottom_right[0]) // 2
    mid_y = (top_left[1] + bottom_right[1]) // 2
    for i in range(top_left[0], bottom_right[0] + 1):
        for j in range(top_left[1], bottom_right[1] + 1):
            if i == j or (i == mid_x and j >= mid_y) or (j == mid_y and i >= mid_x):
                image[i, j] = 1


    # Recursively divide the region into four quadrants
    draw_fractal(image, top_left, (mid_x, mid_y))
    draw_fractal(image, (top_left[0], mid_y + 1), (mid_x, bottom_right[1]))
    draw_fractal(image, (mid_x + 1, top_left[1]), (bottom_right[0], mid_y))
    draw_fractal(image, (mid_x + 1, mid_y + 1), bottom_right)

image = zeros((600, 600))

draw_fractal(image, (0, 0), (599, 599))

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
