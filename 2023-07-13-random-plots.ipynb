```python
# https://stackoverflow.com/a/9295472/271697
from numpy import random
import matplotlib.pyplot as plt

data = random.random((5,5))
img = plt.imshow(data, interpolation='nearest')
img.set_cmap('hot')
plt.axis('off')
```

```python
data = random.random((50,50))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```

```python
data = random.random((500,500))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```

```python
data = random.default_rng().normal(100, 10, size=(100, 100))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```

```python
data = random.default_rng().poisson(10, size=(100, 100))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```

```python
data = random.default_rng().pareto(10, size=(100, 100))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```

```python
data = random.default_rng().pareto(100, size=(100, 100))
plt.imshow(data, interpolation='nearest')
plt.axis('off')
```
